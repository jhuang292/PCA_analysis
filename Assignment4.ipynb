```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('breast-cancer-data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['diagnosis'].unique()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df['diagnosis'])
df['diagnosis'] = le.transform(df['diagnosis']) 
df['diagnosis'].head()
```

```python
# Generate training and test set
X, y = df.iloc[:, 2:].values, df['diagnosis'].values
```

```python
# Standardize the training data
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
X;
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, df[['diagnosis']]], axis = 1)
```

```python
finalDf.head()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [0, 1]
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['diagnosis'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
finalDf['diagnosis'].value_counts()
```

```python
finalDf.columns
```

```python
from sklearn.model_selection import train_test_split
X, y = finalDf[['principal component 1', 'principal component 2']], finalDf[['diagnosis']]
X_train, X_test, y_train, y_test = train_test_split(
         X, y, test_size=0.33, random_state=42)

from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, solver='lbfgs',
                          multi_class='multinomial').fit(X_train, y_train)
print("The accuracy of training set is: ", clf.score(X_train, y_train))
print("The accuracy of test set is: ", clf.score(X_test, y_test))
```
